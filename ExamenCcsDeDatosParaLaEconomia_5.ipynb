```python
# Importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import numpy as np
```

```python
# Cargar el archivo ingresos.txt
df = pd.read_csv("ingresos-1.txt", sep="\t")
```

```python
# Mostrar las primeras filas
print("Primeras filas del dataset:")
display(df.head())
```

```python
# Información general del dataset
print("\nInformación del dataset:")
print(df.info())
```

```python
# Imputar valores faltantes en variables numéricas
df['peso'] = df['peso'].fillna(df['peso'].median())
df['educacion'] = df['educacion'].fillna(df['educacion'].mode()[0])
df['educacion_madre'] = df['educacion_madre'].fillna(df['educacion_madre'].median())
df['educacion_padre'] = df['educacion_padre'].fillna(df['educacion_padre'].median())

# Estandarizar la columna genero
df['genero'] = df['genero'].str.lower()

# Verificar valores únicos en variables categóricas y ordinales
print("Valores únicos en 'camina':", df['camina'].unique())
print("Valores únicos en 'genero':", df['genero'].unique())
```

1. Imputación de valores faltantes:
Se revisaron las columnas numéricas que tenían valores faltantes (peso, educacion, educacion_madre, educacion_padre) y se completaron con medidas estadísticas:
- Mediana: Se utilizó en las columnas donde los valores podían variar considerablemente, como peso, educacion_madre y educacion_padre, ya que la mediana es menos sensible a valores extremos.
- Moda: Se aplicó en la columna educacion, al tratarse de una variable que puede tener valores repetidos y donde el más frecuente es una buena representación.

2. Estandarización de texto:
Se uniformizó el formato de la columna genero al convertir todos los valores a minúsculas. Esto garantiza consistencia y evita duplicados debidos a diferencias de formato como "Masculino" y "masculino".

3. Verificación de valores únicos:

- Se revisaron los valores únicos de las columnas camina y genero para asegurarse de que las categorías fueran coherentes.

- La columna camina mostró valores esperados en su rango, confirmando que no existen valores fuera de lugar.

- La columna genero solo contiene "masculino" y "femenino", sin errores ni inconsistencias.

```python
# Estadísticas descriptivas
print("\nResumen estadístico:")
display(df.describe())
```

```python
# Visualización de la distribución de ingresos
plt.figure(figsize=(8, 5))
sns.histplot(df['ingresos'], bins=30, kde=True)
plt.title("Distribución de ingresos")
plt.xlabel("Ingresos")
plt.ylabel("Frecuencia")
plt.show()
```

La distribución de ingresos presenta un sesgo positivo significativo, donde la mayoría de los valores se concentran entre 0 y 50,000 dólares, mientras que una pequeña cantidad de observaciones alcanza valores extremadamente altos (hasta 400,000 dólares). Este comportamiento puede distorsionar análisis estadísticos y modelos predictivos, ya que los valores extremos (outliers) influyen en las medidas de tendencia central y dispersión.

```python
# Aplicar la transformación logarítmica (sumamos 1 para evitar log(0))
df['log_ingresos'] = np.log1p(df['ingresos'])
```

```python
# Visualizar la nueva distribución
plt.figure(figsize=(8, 5))
sns.histplot(df['log_ingresos'], bins=30, kde=True, color="lightgreen")
plt.title("Distribución logarítmica de ingresos")
plt.xlabel("Log(Ingresos)")
plt.ylabel("Frecuencia")
plt.show()
```

La aplicación de la transformación logarítmica a la variable ingresos ha logrado reducir significativamente el sesgo positivo observado en la distribución original. Ahora, la mayoría de los valores se concentran entre 8 y 11 en escala logarítmica, lo que corresponde aproximadamente a ingresos en rangos medios cuando se revierte la transformación. Además, los valores extremos siguen existiendo pero tienen menor impacto visual y estadístico.

```python
# Contar el número de registros con ingresos cero
num_ceros = (df['ingresos'] == 0).sum()
print(f"Número de registros con ingresos cero: {num_ceros}")

# Porcentaje de registros con ingresos cero
porcentaje_ceros = (num_ceros / len(df)) * 100
print(f"Porcentaje de registros con ingresos cero: {porcentaje_ceros:.2f}%")
```

El número de registros con ingresos cero es 187, representando un 10.30% del total de datos. Dado que este porcentaje es considerable, eliminar estos registros podría llevar a la pérdida de información valiosa.

```python
# Crear una columna binaria para identificar ingresos cero
df['ingresos_cero'] = (df['ingresos'] == 0).astype(int)

# Imputar un valor pequeño (1% del ingreso mínimo no cero) a los registros con ingresos cero
ingreso_minimo = df['ingresos'][df['ingresos'] > 0].min() * 0.01
df['ingresos'] = df['ingresos'].replace(0, ingreso_minimo)

print(f"Se reemplazaron los ingresos cero con: {ingreso_minimo:.2f}")
```

Se han identificado 187 registros con ingresos igual a cero, representando el 10.3% del total. Para garantizar la integridad del análisis y evitar la pérdida de datos, se implementaron dos acciones clave:

1. Nueva columna binaria ingresos_cero:
Esta columna permite distinguir entre registros con ingresos cero y aquellos con ingresos mayores a cero, lo cual es útil para análisis segmentados o modelos predictivos.

2. Imputación de un valor pequeño:
Los ingresos cero fueron reemplazados con el 1% del ingreso mínimo no nulo (equivalente a 2.00). Esto asegura que:

- Las transformaciones logarítmicas puedan aplicarse correctamente.
- Los registros con ingresos cero permanezcan en el análisis sin distorsionar significativamente los resultados.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['educacion_madre', 'educacion_padre']])
plt.title("Boxplot de educación madre y padre")
plt.show()
```

El boxplot revela la presencia de valores atípicos (outliers) en las variables educacion_madre y educacion_padre. Estos valores extremos se encuentran alrededor de 99 años, lo cual es poco realista para años de educación.

```python
median_madre = df['educacion_madre'].median()
median_padre = df['educacion_padre'].median()

df['educacion_madre'] = df['educacion_madre'].apply(lambda x: median_madre if x > 20 else x)
df['educacion_padre'] = df['educacion_padre'].apply(lambda x: median_padre if x > 20 else x)
```

Se reemplazaron los valores atípicos mayores a 20 años en las columnas educacion_madre y educacion_padre con la mediana correspondiente de cada columna. Esta decisión permite mantener los registros en el dataset sin eliminar información, al mismo tiempo que se reduce el impacto de valores extremos en el análisis estadístico. La mediana es una medida robusta que asegura que los datos sean más realistas y consistentes, mejorando así la calidad del análisis posterior.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='ejercicio', y='tenso', data=df, hue='genero', palette="muted")
plt.title("Ejercicio vs Días tensos")
plt.xlabel("Días de ejercicio")
plt.ylabel("Días tensos")
plt.show()
```

```python
# Calcular correlación entre ejercicio y tenso para todo el dataset
correlacion_total = df[['ejercicio', 'tenso']].corr().iloc[0, 1]
print(f"Correlación total entre ejercicio y días tensos: {correlacion_total:.2f}")

# Correlación por género
print("Correlación por género:")
for genero in df['genero'].unique():
    correlacion = df[df['genero'] == genero][['ejercicio', 'tenso']].corr().iloc[0, 1]
    print(f"- {genero.capitalize()}: {correlacion:.2f}")
```

A nivel global, la correlación es apenas 0.01, lo que indica que prácticamente no existe una relación lineal significativa entre estas dos variables en el conjunto de datos analizado.

Al segmentar los resultados por género, se observa que la correlación es igualmente débil:

- Para hombres, la correlación es 0.06, lo cual sugiere una relación mínima y posiblemente aleatoria.
- Para mujeres, la correlación es aún más baja, con un valor de 0.04, lo que refuerza la conclusión de que el número de días de ejercicio no está relacionado con la cantidad de días tensos reportados.

```python
# Verificar valores nulos restantes
print(df.isnull().sum())
```

```python
# Reemplazar valor nulo en 'fumador' con la moda
df['fumador'].fillna(df['fumador'].mode()[0], inplace=True)
```

```python
# Reemplazar valores nulos en 'tenso' y 'malhumorado' con la mediana
df['tenso'].fillna(df['tenso'].median(), inplace=True)
df['malhumorado'].fillna(df['malhumorado'].median(), inplace=True)
```

```python
# Verificar nuevamente los valores nulos
print("Valores nulos después del reemplazo:")
print(df.isnull().sum())
```

```python
# Verificar outliers usando el método IQR solo en columnas numéricas
numerical_df = df.select_dtypes(include=np.number)  # Seleccionar solo columnas numéricas

Q1 = numerical_df.quantile(0.25)
Q3 = numerical_df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((numerical_df < (Q1 - 1.5 * IQR)) | (numerical_df > (Q3 + 1.5 * IQR))).sum()
print("Cantidad de outliers por columna:")
print(outliers)
```

```python
from scipy.stats import mstats

# Aplicar winsorizing para ingresos
df['ingresos'] = mstats.winsorize(df['ingresos'], limits=[0.01, 0.01])
df['log_ingresos'] = mstats.winsorize(df['log_ingresos'], limits=[0.01, 0.01])
```

```python
# Reemplazar outliers en 'peso' usando la mediana
median_peso = df['peso'].median()
df['peso'] = df['peso'].apply(lambda x: median_peso if x > Q3['peso'] + 1.5 * IQR['peso'] else x)
```
